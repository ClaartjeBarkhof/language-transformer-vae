## Gather & display mean of stats for all runs

```python
pd.set_option('display.float_format', '{:.2f}'.format)

VAL_BATCHES = 10
IW_SAMPLES = 200

all_stats = {}
for exp_name in RUN_DIRS:
    all_stats[exp_name] = {}
    
    run_overview = read_overview_csv(exp_name=exp_name)
    
    for row_i, row in run_overview.iterrows():
        run_name, clean_name = row["run_name"], row["clean_name"]
        
        hp_details = parse_run_name(run_name, exp_name)
        
        try:
            sum_stats = get_sum_stats_run(run_name, exp_name=exp_name, val_batches=VAL_BATCHES, iw_samples=IW_SAMPLES)

            for s in list(sum_stats.keys()):
                if torch.is_tensor(sum_stats[s]):
                    if sum_stats[s].dim() > 1:
                        # print(s, sum_stats[s].shape)
                        # del sum_stats[s]
                        
                        # don't delete, just consider the first sample
                        sum_stats[s] = sum_stats[s][:, 0].mean().item()
                    else:
                        sum_stats[s] = sum_stats[s].mean().item()
                else:
                    sum_stats[s] = np.mean(sum_stats[s])
            all_stats[exp_name][clean_name] = {**sum_stats, **hp_details}
            all_stats[exp_name][clean_name]['ppl'] = np.exp(-all_stats[exp_name][clean_name]["iw_ll_p_w"])
        
        except Exception as e:
            print(f"** Error for {clean_name}, {e}")

all_stats_dfs = {}
for exp_name, exp_stats in all_stats.items():
    print("*"*80, "\n", exp_name, "\n", "*"*79)
    df_exp = pd.DataFrame(exp_stats).transpose()
    all_stats_dfs[exp_name] = df_exp
    print(df_exp.columns)
    display(df_exp)
```

# Compare results along a specific hyper parameter axis

```python
def compare_along_hp_axis(df, compare_axis, select_metrics, a):
    # Global information
    all_hp_axes = ["dataset", "drop", "optimisation", "mech_string"]
    assert compare_axis in all_hp_axes, f"compare axis should be part of hp axes: {all_hp_axes}"

    other_hp_axes = [a for a in all_hp_axes if a != compare_axis]

    # print("Compare axis:", compare_axis)
    # print("Other axes:", other_hp_axes)
    # print("All hp axes:", all_hp_axes)

    # Change the order to: other hp, compare axis, selected metrics
    order = other_hp_axes + [compare_axis]
    other_cols = list(set(df.columns) - set(order))
    full_order = order + other_cols

    df = df[full_order].sort_values(order) # order on [other hp, compare axis]

    include_hp_explicit = True
    if include_hp_explicit:
        select_metrics += other_hp_axes

    # Make multi-level index
    df = df.set_index([df.index, compare_axis])[select_metrics]
    df.index = df.index.rename([None, f'Axis of comparison: {compare_axis}'])
    return df
    
# Input
df = all_stats_dfs["Runs"]
compare_axis = "mech_string"
select_metrics = ["iw_ll", "iw_ll_x_gen", "log_p_z", "log_q_z_x", "log_q_z"]

compare_along_hp_axis(df, compare_axis, select_metrics)
```
